```python
class Yeuem:
    def __init__(self, name, nangluong):
        self.name = name
        self.nangluong = nangluong

    def tancong(self, target):
        self.nangluong = self.nangluong + target
em = Yeuem("Em", 100)
em.tancong(10)
print(f"nang luong con lai: {em.nangluong}, ten em: {em.name}")
```
